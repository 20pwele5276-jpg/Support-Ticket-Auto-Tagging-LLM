```python
# =========================
# TASK 5 - AUTO TAGGING SUPPORT TICKETS
# LLM Prompt Engineering
# =========================

import pandas as pd

# =========================
# SAMPLE SUPPORT TICKETS
# =========================

tickets = [
    "My internet is not working since morning",
    "I want to reset my password but not receiving email",
    "Payment failed but money deducted from account",
    "App is crashing when I open it",
    "How can I upgrade my subscription plan?"
]

# =========================
# SIMPLE TAGGING FUNCTION (LLM STYLE SIMULATION)
# =========================

def tag_ticket(ticket):
    prompt = f"""
You are a support ticket classifier.
Classify the ticket into one category:
- Technical Issue
- Billing Issue
- Account Issue
- App Bug
- General Query

Ticket: {ticket}

Return only the category.
"""

    # Simulated LLM response logic (fast version)
    if "internet" in ticket.lower() or "crash" in ticket.lower():
        return "Technical Issue"
    elif "payment" in ticket.lower() or "money" in ticket.lower():
        return "Billing Issue"
    elif "password" in ticket.lower() or "account" in ticket.lower():
        return "Account Issue"
    elif "upgrade" in ticket.lower():
        return "General Query"
    else:
        return "General Query"

# =========================
# TEST MODEL
# =========================

for t in tickets:
    print("Ticket:", t)
    print("Predicted Tag:", tag_ticket(t))
    print("-"*40)
```
